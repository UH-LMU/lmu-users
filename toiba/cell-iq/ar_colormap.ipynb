```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("/work/data/mushtaq/test_well_output/ar/2020_06_29_16_19_09_43_segmentation_ar.tif")
```

```python
img = img/10
```

```python
import copy

fig, ax = plt.subplots()
#fig.figsize = (1392,1040)
cmap = copy.copy(matplotlib.cm.get_cmap("jet"))
im = ax.imshow(img, cmap=cmap, vmin=1, vmax=5)
#ax.axis('off')
fig.colorbar(im, ax=ax)
#plt.show()
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

plt.figure()
ax = plt.gca()
im = ax.imshow(img, cmap=cmap, vmin=1, vmax=5)

divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.1)
plt.colorbar(im,cax=cax)
```
